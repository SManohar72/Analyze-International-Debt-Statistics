```python
## reading the csv file
df1 = pd.read_csv('../modified data-sets/IDS_ALLCountries_Data-1.csv')
df1.head()
```

```python
#connect to a database
conn = sqlite3.connect("all_countries.db") 
```

```python
#store your table in the database:
df1.to_sql('all_countries', conn)
```

## Test Query

```python
query = "select * from all_countries"
ans1 = pd.read_sql(query, conn)
ans1.head()
```

## 2) Finding the number of distinct countries

```python
query2 = 'select distinct Country_Name from all_countries'
ans2 = pd.read_sql(query2, conn)
```

```python
ans2.head()
```

```python
print("Number of Distinct Countries : ",len(ans2))
```

## 3) Finding out the distinct debt indicators

```python
query3 = 'select distinct Country_Code as "Distinct Dept Indicator" from all_countries'
ans3 = pd.read_sql(query3, conn)
```

```python
ans3.head()
```

```python
print("Number of Distinct Dept Indicators : ",len(ans3))
```
